```python
import pandas as pd
```

```python
load_dataset = pd.read_excel("Pre-ProcessedbeforeScalingandEncoding_OL_ST_R.xlsx")
```

```python
load_dataset.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
numerical_cols = load_dataset.select_dtypes(include=['int64', 'float64']).columns.tolist()
scaler = MinMaxScaler()

scaled_array = scaler.fit_transform(load_dataset[numerical_cols])
scaled_dataset = pd.DataFrame(scaled_array, columns=numerical_cols, index=load_dataset.index)
load_dataset[numerical_cols] = scaled_dataset
```

```python
load_dataset.head()
```

```python
scaled_dataset.head()
```

```python
load_dataset = pd.get_dummies(load_dataset, columns=['Unit', 'Category'], drop_first=True)
```

```python
load_dataset.head()
```

```python
load_dataset = load_dataset.astype({col: 'int' for col in load_dataset.select_dtypes('bool').columns})
```

```python
load_dataset.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
load_dataset['Drug_Name_Label'] = label_encoder.fit_transform(load_dataset['Drug Name'])
```

```python
load_dataset.to_excel("Pre-ProcessedafterScalingandEncoding_OL_ST_R.xlsx", index=False)
```
